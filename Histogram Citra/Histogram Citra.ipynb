EKUALISAI HISTOGRAM CITRA

```python
# Import Library untuk Histogram Citra
import numpy as np 
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
img = cv.imread('manycats.jpg',0)       #membaca gambar
cv.imshow("Citra Overexposure", img)    #menampilkan gambar
cv.waitKey(0)                           #menunggu user input key
cv.destroyAllWindows()                  #menutup semua window
```

```python
# PERHITUNGAN NORMALISAI CUMULATIVE DENSITIY FUNCTION
hist,bins = np.histogram(img.flatten(),256,[0,256])
cdf = hist.cumsum()
cdf_normalized = cdf * float(hist.max()) / cdf.max()
```

```python
#melakukan plotting histogram dan hasil kalkulasi CDF
plt.plot(cdf_normalized, color = 'g')
plt.hist(img.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
plt.legend(('cdf','histogram'), loc = 'upper left')
plt.show()
```

```python
#masking CDF pada histogram gambar
cdf_m = np.ma.masked_equal(cdf,0)
cdf_m = (cdf_m - cdf_m.min())*255/(cdf_m.max()-cdf_m.min())
cdf = np.ma.filled(cdf_m,0).astype('uint8')
```

```python
#Masking hasil CDF ke gambar
img2 = cdf[img]
cv.imshow("Citra Hasil Histogram Equalization", img2)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
#plotting gambar
hist2,bins2 = np.histogram(img2.flatten(),256,[0,256])
cdf2 = hist2.cumsum()
cdf_normalized2 = cdf2 * float(hist2.max()) / cdf2.max()

plt.plot(cdf_normalized2, color = 'b')
plt.hist(img2.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
plt.legend(('cdf','histogram'), loc = 'upper left')
plt.show()
```
